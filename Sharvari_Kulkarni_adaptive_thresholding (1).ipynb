# ADAPTIVE THRESHOLDING

The method in which the threshold values are calculated separately for separate smaller regions is called as 
adaptive thresholding.
This means that using adaptive thresholding we can determine the threshold value for a pixel based on small region around it.
So we get different threshold value for different regions.

OpenCV provides Adaptive threshold operation using cv2.adaptiveThreshold() method.

Syntax:
    
    cv2.adaptiveThreshold(src, dst, maxValue, adaptiveMethod, thresholdType, blockSize, C)












src name of the image on which we wish to perform thresholding operation . This image should be grayscale.

dst output array of the same size and type and the same number of channels as src.

maxval maximum value which is assigned to pixel values exceeding the threshold

adaptiveMethod: adaptiveMethod decides how the threshold value is calculated:

cv2.ADAPTIVE_THRESH_MEAN_C: The threshold value is the mean of the neighbourhood area minus the constant C.
cv2.ADAPTIVE_THRESH_GAUSSIAN_C: The threshold value is a gaussian-weighted sum of the neighbourhood values minus the constant C.

thresholdType: Representing the type of threshold to be used.

blockSize: It determines the size of the neighbourhood area.

C: C is a constant that is subtracted from the mean or weighted sum of the neighborhood pixels.

```python
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("img_adaptivethreshold.jpg")
plt.figure(figsize=(20,8))
plt.imshow(img)
```

CONVERTING A COLOURED IMAGE INTO A GRAY SCALE IMAGE

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img_gray,cmap='gray')
```

BINARY IMAGE OF THE ABOVE GRAY SCALE IMAGE

```python
_,th1=cv2.threshold(img_gray,150,255,cv2.THRESH_BINARY)
plt.figure(figsize=(20,8))
plt.imshow(th1,cmap='gray')
```

150 is the threshold value used.But we are not able to see the perfect image as the threshold is only calculated for whole image and 
the illumination is not similar in all regions.

Hence for better visualization we will go with adaptive thresholding.

In adaptive thresholding the threshold value depends on the pixels of the surrounding .

# ADAPTIVE MEAN THRESHOLDING

```python
th2=cv2.adaptiveThreshold(img_gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
plt.figure(figsize=(20,8))
plt.imshow(th2,cmap='gray')
```

Here 11 specifies the number of pixels on which the threshold will be applied that is neighbourhood area or block size and 
2 is the value that needs to be subtracted from thethreshold value that is the constant C .

# ADAPTIVE GAUSSIAN THRESHOLDING

```python
th3=cv2.adaptiveThreshold(img_gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,15,3)
plt.figure(figsize=(20,8))
plt.imshow(th3,cmap='gray')
```

Here 15 is the  number of pixels on which the threshold will be applied that is neighbourhood area or block size and 3 is the 
value that needs to be subtracted from thethreshold value that is the constant C .
